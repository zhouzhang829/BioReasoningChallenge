```python
import pandas as pd   
import json
import numpy as np 
from pathlib import Path
import subprocess
import json 
#sk-lit-93cea5ba-7b93-474c-82e9-109b4095d708
# Load data
train = pd.read_csv("prompt_a/train.csv")
test = pd.read_csv("prompt_a/test.csv")
# Load prompt template
SEEDS = [42, 43, 44]

results = []

# Read template once
prompt_template = Path("prompt1.txt").read_text()

for _, row in train.head(10).iterrows():

    # Fill the template
    prompt = prompt_template.format(
        model_id=row["id"],
        pert=row["pert"],
        gene=row["gene"],
    )

    # Write a temporary prompt file
    Path("prompt_tmp.txt").write_text(prompt)

    subprocess.run(
        [
            "python",
            "gpt120_openai.py",
            "prompt_tmp.txt",
            "output.txt",
            "42",
        ],
        check=True,
    )

    with open("response.json", "r") as f:
        response = json.load(f)
    message = response["choices"][0]["message"]
    results.append({
        "model_id": row["id"],
        "pert": row["pert"],
        "gene": row["gene"],
        "content": message.get("content", ""),
        "reasoning_content": message.get("reasoning_content", ""),
    })
results = pd.DataFrame(results)
results.to_csv("gpt120_results.csv", index=False)

```
